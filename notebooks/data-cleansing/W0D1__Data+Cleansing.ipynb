# <center> **Data Cleansing** 

---

> A process of detecting and rectifying (or deleting) of untrustworthy, inaccurate or outdated information from a data set, archives, table, or database. It helps you to identify incomplete, incorrect, inaccurate or irrelevant parts of the data.

> When combining multiple data sources, there are many opportunities for data to be duplicated or mislabeled. If data is incorrect, outcomes and algorithms are unreliable, even though they may look correct.

```python
# 10k samples from SF DBI permits record
import pandas as pd
df = pd.read_csv("Sample10k_Building_Permits.csv", sep=',', header=0)
df.shape
```

```python
# Exact instance duplicated?
df.duplicated().sum()
```

```python
# Individual column
df.duplicated(subset='Permit Number').sum()
```

```python
df['Permit Number'].value_counts()
```

```python
df['Permit Number'].value_counts().nlargest(1)
```

```python
# DFobject[condition]
# Here condition --> DFobject["Column Name"] == <a_legit_value>
df[df['Permit Number'] == '201602179765']
```

```python
# DFobject.loc[condition, <single_column_name_or_a_list_of_columns>]
df.loc[df['Permit Number'] == '201602179765', ['Permit Number', 'Permit Creation Date']]
```

```python
temp = df.drop_duplicates(subset='Permit Number', keep='first')
temp.shape
```

```python
temp[temp['Permit Number'] == '201602179765']
```

```python
temp = df.drop_duplicates(subset='Permit Number', keep='last')
temp.shape
```

```python
temp[temp['Permit Number'] == '201602179765']
```

```python
temp = df.drop_duplicates(subset='Permit Number', keep=False)
temp.shape
```

```python
temp[temp['Permit Number'] == '201602179765']
```

```python
# delete an object
del temp
```

```python
# Multi-level keys
df.duplicated(subset=['Permit Number', 'Record ID']).sum()
```
